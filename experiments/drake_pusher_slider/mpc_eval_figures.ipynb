```python
c = np.array([(255, 105, 65), (0, 69, 255), (50, 205, 50), (0, 165, 255), (238, 130, 238),
             (128, 128, 128), (30, 105, 210), (147, 20, 255), (205, 90, 106), (255, 215, 0)])

c = c *1/255.0


# colors for each model
colors_dict = {}
colors_dict['GT 3D'] = c[0]
colors_dict['SDS'] = c[1]
colors_dict['WDS'] = c[2]
colors_dict['transporter 2D'] = c[3]
colors_dict['transporter 3D'] = c[4]
colors_dict['autoencoder'] = c[5]

add_error_bars = False


# maps pandas dataframe name to display name
exp_names_dict = OrderedDict()
exp_names_dict['correlle_mug-small_many_colors_600'] = "mugs"
exp_names_dict['box_on_side'] = "box on side (camera angled)"
exp_names_dict['box_push_1000_angled'] = "box push (camera angled)"
exp_names_dict['box_push_1000_top_down'] = "box push (camera top down)"


baseline_model_types = dict()
d = baseline_model_types
d['object_delta'] = "goal states"
d['DD_3D_spatial'] = "SDS"
d['DD_3D_all_z'] = "WDS"
d['transporter_3D'] = "transporter 3D"
# d['transporter_3D_T_aug'] = "transporter 3D T_aug" # actually results in worse performance
d['transporter_2D'] = "transporter 2D"
d['autoencoder_z_dim_64'] = "autoencoder"



nrows = len(exp_names_dict)
ncols = 2
figsize = (6.4*2, 4.8*nrows)
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, squeeze=False, figsize=figsize)

axes_row_counter = 0


for exp_name, exp_name_display in exp_names_dict.items():
    
    print("exp_name_display", exp_name_display)
    df_exp = df_full[df_full['dataset_name'] == exp_name]
    
    pos = []
    pos_std_dev = []
    angle = []
    angle_std_dev = []
    method_names = []
    colors_list = []
    
    
    for model_name, display_name in baseline_model_types.items():
        if model_name == "object_delta":
            continue
        df = df_exp[df_exp['model_name'] == model_name]
        
        pos.append(df['position_error'].mean()*100)
        pos_std_dev.append(df['position_error'].std()*100)
        
        angle.append(df['angle_error_degrees'].mean())
        angle_std_dev.append(df['angle_error_degrees'].std())
        
        method_names.append(display_name)
        
        colors_list.append(colors_dict[display_name])
        
    
    
    
    # position
    ax = axes[axes_row_counter,0]
    ax.set_title(exp_name_display)
    plt.sca(ax)
    
    
    yerr = pos_std_dev if add_error_bars else None
    x_pos = [i for i, _ in enumerate(pos)]
    plt.bar(x_pos, pos, color=colors_list, yerr=yerr)
    plt.ylabel("Position Error (cm)")
    plt.xticks([])  # Disable xticks.
    
    if axes_row_counter == (nrows-1):
        plt.xticks(x_pos, method_names)
        plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
    
    
    # angle
    ax = axes[axes_row_counter,1]
    ax.set_title(exp_name_display)
    plt.sca(ax)
    yerr = angle_std_dev if add_error_bars else None
    plt.bar(x_pos, angle, color=colors_list, yerr=yerr)
    plt.ylabel("Angle Error (degrees)")
    plt.xticks([])  # Disable xticks.
    
    if axes_row_counter == (nrows-1):
        plt.xticks(x_pos, method_names)
        plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
    
    
    axes_row_counter += 1
    
    
save_dir = os.path.join(get_project_root(), 'sandbox/paper_figures/results')

if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    
save_file = os.path.join(save_dir, 'baseline_results.png')
fig.savefig(save_file)
plt.show()

    
    
    
        
    
        


```

```python
c = np.array([(255, 105, 65), (0, 69, 255), (50, 205, 50), (0, 165, 255), (238, 130, 238),
             (128, 128, 128), (30, 105, 210), (147, 20, 255), (205, 90, 106), (255, 215, 0)])

c = c *1/255.0


# colors for each model
colors_dict = {}
# colors_dict['GT 3D'] = c[0]
colors_dict['GT 3D'] = [1,0,0]
colors_dict['SDS'] = c[1]
colors_dict['WDS'] = c[2]
colors_dict['DS'] = c[3]
colors_dict['WSDS'] = c[4]
colors_dict['SDS no T_aug'] = c[5]
colors_dict['GT 3D no T_aug'] = c[6]
colors_dict['WDS no T_aug'] = c[7]



# colors_dict['transporter 2D'] = c[3]
# colors_dict['transporter 3D'] = c[4]
# colors_dict['autoencoder'] = c[5]

add_error_bars = False


# maps pandas dataframe name to display name
exp_names_dict = OrderedDict()
exp_names_dict['correlle_mug-small_many_colors_600'] = "mugs"
exp_names_dict['box_on_side'] = "box on side (camera angled)"
exp_names_dict['box_push_1000_angled'] = "box push (camera angled)"
exp_names_dict['box_push_1000_top_down'] = "box push (camera top down)"





ablation_model_types = OrderedDict()
# ablation_model_types['object_delta'] = "goal states"
ablation_model_types['GT_3D'] = "GT 3D"
ablation_model_types['DD_3D_spatial'] = "SDS"
ablation_model_types['DD_3D_all_z'] = "WDS"
ablation_model_types["DD_3D_all"] = "DS"
ablation_model_types['DD_3D_spatial_z'] = "WSDS"
ablation_model_types['DD_3D_spatial_no_T_aug'] = "SDS no T_aug"
ablation_model_types['GT_3D_no_T_aug'] = "GT 3D no T_aug"
ablation_model_types['DD_3D_all_z_no_T_aug'] = "WDS no T_aug"



nrows = len(exp_names_dict)
ncols = 2
figsize = (6.4*2, 4.8*nrows)
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, squeeze=False, figsize=figsize)

axes_row_counter = 0


for exp_name, exp_name_display in exp_names_dict.items():
    
    print("exp_name_display", exp_name_display)
    df_exp = df_full[df_full['dataset_name'] == exp_name]
    
    pos = []
    pos_std_dev = []
    angle = []
    angle_std_dev = []
    method_names = []
    colors_list = []
    
    
    for model_name, display_name in ablation_model_types.items():
        if model_name == "object_delta":
            continue
        df = df_exp[df_exp['model_name'] == model_name]
        
        if model_name == "GT_3D" and exp_name_display == "mugs":
            print("adding one-off for GT_3D with mugs")
            pos.append(0)
            pos_std_dev.append(0)
            angle.append(0)
            angle_std_dev.append(0)
            display_name = ablation_model_types['GT_3D']
            method_names.append(display_name)
            colors_list.append(colors_dict[display_name])
        else:

            pos.append(df['position_error'].mean()*100)
            pos_std_dev.append(df['position_error'].std()*100)
            angle.append(df['angle_error_degrees'].mean())
            angle_std_dev.append(df['angle_error_degrees'].std())
            method_names.append(display_name)

            colors_list.append(colors_dict[display_name])
        
    
    print("\n\n-----%s-----" %(exp_name_display))
    print("len(pos)", len(pos))
    print("len(angle)", len(angle))
    
    # position
    ax = axes[axes_row_counter,0]
    ax.set_title(exp_name_display)
    plt.sca(ax)
    
    
    yerr = pos_std_dev if add_error_bars else None
    x_pos = [i for i, _ in enumerate(pos)]
    plt.bar(x_pos, pos, color=colors_list, yerr=yerr)
    plt.ylabel("Position Error (cm)")
    plt.xticks([])  # Disable xticks.
    
    if axes_row_counter == (nrows-1):
        plt.xticks(x_pos, method_names)
        plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
    
    
    # angle
    ax = axes[axes_row_counter,1]
    ax.set_title(exp_name_display)
    plt.sca(ax)
    yerr = angle_std_dev if add_error_bars else None
    plt.bar(x_pos, angle, color=colors_list, yerr=yerr)
    plt.ylabel("Angle Error (degrees)")
    plt.xticks([])  # Disable xticks.
    
    if axes_row_counter == (nrows-1):
        plt.xticks(x_pos, method_names)
        plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
    
    
    axes_row_counter += 1
    
    
save_dir = os.path.join(get_project_root(), 'sandbox/paper_figures/results')

if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    
save_file = os.path.join(save_dir, 'ablation_results.png')
fig.savefig(save_file)
plt.show()

    
    
    
        
    
        


```
